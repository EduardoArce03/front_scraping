```python
import pickle
import time
import random
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException

driver = webdriver.Chrome()

def cargar_cookies_x():
    driver.get("https://x.com")
    time.sleep(3)
    try:
        with open("x_cookies.pkl", "rb") as file:
            cookies = pickle.load(file)
            for cookie in cookies:
                driver.add_cookie(cookie)
        driver.refresh()
        time.sleep(5)
    except FileNotFoundError:
        print("❌ Error: No se encontró el archivo de cookies.")

def realizar_scraping_x(tema):
    # 1. Ir a la búsqueda (Filtrando por los más recientes/populares)
    search_url = f"https://x.com/search?q={tema}&src=typed_query"
    driver.get(search_url)
    print(f"Buscando en X: {tema}...")
    time.sleep(7)

    # 2. Abrir CSV
    with open('comentarios_x_anonimos.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Post_Texto"])

        # 3. Scroll Infinito para cargar contenido
        # En X, los posts cargan conforme bajas.
        for i in range(5): # Bajamos 5 veces para recolectar suficiente data
            print(f"Haciendo scroll {i+1}...")
            
            # Buscamos los posts usando el data-testid que es estandar en X
            tweets = driver.find_elements(By.CSS_SELECTOR, 'div[data-testid="tweetText"]')
            
            for tweet in tweets:
                try:
                    texto = tweet.text.strip().replace('\n', ' ')
                    # Solo guardamos si tiene contenido real para la investigación
                    if len(texto) > 10:
                        writer.writerow([texto])
                        print(f"  [+] Tweet extraído: {texto[:50]}...")
                except:
                    continue
            
            # Scroll hacia abajo
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(random.uniform(4, 6))

    print("\n✅ Proceso completado en X. Revisa 'comentarios_x_anonimos.csv'")

# --- Ejecución ---
cargar_cookies_x()
realizar_scraping_x("IA desarrollo software")
```
